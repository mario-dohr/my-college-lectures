```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(-60,60,1000)
y = np.random.uniform(-40,40,1000)
Y = np.vstack((x,y)).T
plt.figure(figsize=(10, 10))
plt.scatter(x,y,)
Y.shape
```

```python
alpha = np.pi/6
U = np.array([[np.cos(alpha),-np.sin(alpha)], [np.sin(alpha),np.cos(alpha)]])
U
```

```python
X = np.matmul(U,Y.T).T
X.shape
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(X[:,0],X[:,1])
```

# Fast ICA with skitlearn

```python
from sklearn.decomposition import FastICA
transformer = FastICA(random_state=0,whiten='unit-variance')
Y_2 = transformer.fit_transform(X)
```

```python
Y_2.shape
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(Y_2[:,1],Y_2[:,0])
```

```python
alpha = -np.pi/2
U = np.array([[np.cos(alpha),-np.sin(alpha)], [np.sin(alpha),np.cos(alpha)]])

Y = np.matmul(U,Y_2.T).T
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(Y[:,1],Y[:,0])
```
